```python
import pandas as pd
import random
from datetime import timedelta

random.seed(42)
```

```python
companies_df = pd.read_csv(
    "projectflow-analytics/data/raw/companies.csv"
)

users_df = pd.read_csv(
    "projectflow-analytics/data/raw/users.csv"
)


companies_df.head()
```

```python
project_names = [
    "Customer Analytics Platform",
    "AI Chat Assistant",
    "Mobile App Redesign",
    "Payment Integration System",
    "HR Management Portal",
    "Marketing Automation Tool",
    "Sales Dashboard",
    "Cloud Migration Project",
    "Security Monitoring System",
    "Data Warehouse Setup"
]


project_status = [
    "Planning",
    "Active",
    "Completed",
    "On Hold",
    "Cancelled"
]


priority = [
    "Low",
    "Medium",
    "High",
    "Critical"
]
```

```python
projects = []

project_id = 1


for company in companies_df.itertuples():


    company_users = users_df[
        users_df["company_id"] == company.company_id
    ]


    total_projects = random.randint(5, 15)


    for i in range(total_projects):


        start_date = (
            pd.to_datetime(company.registration_date)
            +
            timedelta(days=random.randint(30,500))
        )


        projects.append({

            "project_id": project_id,


            "project_code": f"PRJ{project_id:04}",


            "company_id": company.company_id,


            "created_by": random.choice(
                company_users["user_id"].tolist()
            ),


            "project_name": random.choice(project_names),


            "project_status": random.choices(
                project_status,
                weights=[10,45,35,8,2]
            )[0],


            "priority": random.choices(
                priority,
                weights=[20,50,25,5]
            )[0],


            "start_date": start_date.date(),


            "due_date": (
                start_date
                +
                timedelta(days=random.randint(30,365))
            ).date()

        })


        project_id += 1
```

```python
projects_df = pd.DataFrame(projects)
projects_df.head()
```

```python
projects_df.to_csv(
    "projectflow-analytics/data/raw/projects.csv",
    index=False
)
```
