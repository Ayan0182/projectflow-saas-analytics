```python
import pandas as pd
import random
from datetime import datetime, timedelta

random.seed(42)
```

```python
users_df = pd.read_csv(
    "projectflow-analytics/data/raw/users.csv"
)


users_df.head()
```

```python
activity_types = [
    "login",
    "logout",
    "project_created",
    "task_created",
    "task_completed",
    "comment_added",
    "ai_used"
]
```

```python
activity_logs = []

activity_id = 1


for user in users_df.itertuples():


    # active users have more activity

    if user.user_status == "Active":

        total_activity = random.randint(50,200)

    else:

        total_activity = random.randint(5,30)



    for i in range(total_activity):


        activity_logs.append({

            "activity_id": activity_id,


            "user_id": user.user_id,


            "activity_type": random.choices(

                activity_types,

                weights=[
                    35,   # login
                    20,   # logout
                    5,    # project create
                    15,   # task create
                    15,   # task complete
                    8,    # comments
                    2     # AI
                ]

            )[0],


            "activity_time": (

                datetime(2026,1,1)

                +

                timedelta(

                    days=random.randint(0,180),

                    hours=random.randint(0,23),

                    minutes=random.randint(0,59)

                )

            )

        })


        activity_id += 1
```

```python
activity_logs_df = pd.DataFrame(
    activity_logs
)


activity_logs_df.head()
```

```python
activity_logs_df.to_csv(
    "projectflow-analytics/data/raw/activity_logs.csv",
    index=False
)
```
