```python
import pandas as pd
import random
from datetime import timedelta

random.seed(42)
```

```python
tasks_df = pd.read_csv(
    "projectflow-analytics/data/raw/tasks.csv"
)

users_df = pd.read_csv(
    "projectflow-analytics/data/raw/users.csv"
)


tasks_df.head()
```

```python
comment_messages = [
    "Task completed successfully",
    "Need some improvements",
    "Working on this task",
    "Bug fixed",
    "Code review completed",
    "Testing is in progress",
    "Changes requested",
    "Approved",
    "Waiting for confirmation",
    "Documentation updated"
]
```

```python
comments = []

comment_id = 1


for task in tasks_df.itertuples():


    total_comments = random.randint(0,5)


    for i in range(total_comments):


        comments.append({

            "comment_id": comment_id,


            "task_id": task.task_id,


            "user_id": task.assigned_user_id,


            "comment_text": random.choice(
                comment_messages
            ),


            "commented_at": (
                pd.to_datetime(task.created_date)
                +
                timedelta(
                    days=random.randint(0,20),
                    hours=random.randint(1,23)
                )
            )

        })


        comment_id += 1
```

```python
comments_df = pd.DataFrame(comments)
comments_df.head()
```

```python
comments_df.to_csv(
    "projectflow-analytics/data/raw/comments.csv",
    index=False
)
```
