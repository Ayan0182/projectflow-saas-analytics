```python
import pandas as pd
import random
from datetime import datetime, timedelta

random.seed(42)
```

```python
companies_df = pd.read_csv(
    "projectflow-analytics/data/raw/companies.csv"
)

companies_df.head()
```

```python
first_names = [
    "Aarav",
    "Vivaan",
    "Aditya",
    "Rahul",
    "Samir",
    "Priya",
    "Sneha",
    "Ananya",
    "Kavya",
    "Neha"
]


last_names = [
    "Sharma",
    "Patel",
    "Singh",
    "Mehta",
    "Verma",
    "Shah"
]


roles = [
    "Owner",
    "Project Manager",
    "Developer",
    "Designer",
    "QA"
]
```

```python
users = []

user_id = 1


for company in companies_df.itertuples():

    total_users = random.randint(20, 60)


    for i in range(total_users):

        first = random.choice(first_names)

        last = random.choice(last_names)


        users.append({

            "user_id": user_id,

            "user_code": f"USR{user_id:04}",

            "company_id": company.company_id,

            "full_name": first + " " + last,

            "email": f"{first.lower()}.{last.lower()}{user_id}@projectflow.com",

            "role": random.choices(
                roles,
                weights=[3,10,60,15,12]
            )[0],

            "join_date": (
                pd.to_datetime(company.registration_date)
                +
                timedelta(days=random.randint(1,300))
            ).date(),

            "last_login": datetime(
                2026,
                random.randint(1,7),
                random.randint(1,28),
                random.randint(0,23),
                random.randint(0,59)
            ),

            "user_status": random.choices(
                ["Active","Inactive"],
                weights=[95,5]
            )[0]

        })


        user_id += 1
```

```python
users_df = pd.DataFrame(users)

users_df.head()
```

```python
users_df.to_csv(
    "projectflow-analytics/data/raw/users.csv",
    index=False
)
```
