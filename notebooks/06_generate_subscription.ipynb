```python
import pandas as pd
import random
from datetime import timedelta

random.seed(42)
```

```python
companies_df = pd.read_csv(
    "projectflow-analytics/data/raw/companies.csv"
)

companies_df.head()
```

```python
plans = {
    "free": 0,
    "starter": 999,
    "pro": 4999,
    "enterprise": 19999
}


status = [
    "active",
    "expired",
    "cancelled"
]
```

```python
subscriptions = []

subscription_id = 1


for company in companies_df.itertuples():


    first_plan = random.choice(
        ["free", "starter", "pro"]
    )


    start_date = pd.to_datetime(
        company.registration_date
    )


    subscriptions.append({

        "subscription_id": subscription_id,

        "company_id": company.company_id,

        "plan_name": first_plan,

        "monthly_price": plans[first_plan],

        "start_date": start_date.date(),

        "end_date": None,

        "subscription_status": "active"

    })


    subscription_id += 1



    # some companies upgrade later

    if random.random() < 0.35:


        upgrade_plan = random.choice(
            ["pro", "enterprise"]
        )


        upgrade_date = (
            start_date
            +
            timedelta(
                days=random.randint(90,400)
            )
        )


        subscriptions[-1]["end_date"] = upgrade_date.date()

        subscriptions[-1]["subscription_status"] = "expired"


        subscriptions.append({

            "subscription_id": subscription_id,

            "company_id": company.company_id,

            "plan_name": upgrade_plan,

            "monthly_price": plans[upgrade_plan],

            "start_date": upgrade_date.date(),

            "end_date": None,

            "subscription_status": "active"

        })


        subscription_id += 1
```

```python
subscription_df = pd.DataFrame(
    subscriptions
)

subscription_df.head()
```

```python
subscription_df.to_csv(
    "projectflow-analytics/data/raw/subscription_history.csv",
    index=False
)
```
