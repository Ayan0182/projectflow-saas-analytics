```python
import pandas as pd
import random
from datetime import datetime, timedelta

random.seed(42)
```

```python
company_names = [
    "Nexora Technologies",
    "CloudMint Labs",
    "BrightStack Systems",
    "CodeSphere Tech",
    "DataSprint Solutions",
    "PixelForge Labs",
    "WorkNest Technologies",
    "ElevateSoft",
    "CoreVista Solutions",
    "TaskNova Systems"
]


industries = [
    "SaaS",
    "FinTech",
    "EdTech",
    "HealthTech",
    "HRTech",
    "AI",
    "Cyber Security"
]


locations = {
    "Ahmedabad": "Gujarat",
    "Bengaluru": "Karnataka",
    "Mumbai": "Maharashtra",
    "Pune": "Maharashtra",
    "Hyderabad": "Telangana",
    "Chennai": "Tamil Nadu",
    "Delhi": "Delhi"
}


company_status = [
    "Active",
    "Inactive",
    "Suspended"
]
```

```python
companies = []


for i, company_name in enumerate(company_names, start=1):

    city = random.choice(list(locations.keys()))

    companies.append({

        "company_id": i,

        "company_code": f"CMP{i:03}",

        "company_name": company_name,

        "industry": random.choice(industries),

        "city": city,

        "state": locations[city],

        "employee_count": random.randint(20, 5000),

        "registration_date": (
            datetime(2023,1,1)
            +
            timedelta(days=random.randint(0,1095))
        ).date(),

        "company_status": random.choices(
            company_status,
            weights=[90,8,2]
        )[0]

    })
```

```python
companies_df = pd.DataFrame(companies)
companies_df
```
