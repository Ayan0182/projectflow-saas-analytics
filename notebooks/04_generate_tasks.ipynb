```python
import pandas as pd
import random
from datetime import timedelta

random.seed(42)
```

```python
projects_df = pd.read_csv(
    "projectflow-analytics/data/raw/projects.csv"
)

users_df = pd.read_csv(
    "projectflow-analytics/data/raw/users.csv"
)


projects_df.head()
```

```python
task_names = [
    "Design User Interface",
    "Create Database Schema",
    "Develop API",
    "Testing Module",
    "Fix Bugs",
    "Create Dashboard",
    "Write Documentation",
    "Optimize Performance",
    "Deploy Application",
    "Security Testing"
]


task_status = [
    "pending",
    "in progress",
    "completed"
]


priority = [
    "low",
    "medium",
    "high",
    "critical"
]
```

```python
tasks = []

task_id = 1


for project in projects_df.itertuples():


    company_users = users_df[
        users_df["company_id"] == project.company_id
    ]


    total_tasks = random.randint(10,40)


    for i in range(total_tasks):


        created_date = (
            pd.to_datetime(project.start_date)
            +
            timedelta(days=random.randint(1,30))
        )


        status = random.choices(
            task_status,
            weights=[25,35,40]
        )[0]


        completed_date = None


        if status == "completed":

            completed_date = (
                created_date
                +
                timedelta(days=random.randint(1,20))
            ).date()


        tasks.append({

            "task_id": task_id,

            "task_code": f"TSK{task_id:05}",

            "project_id": project.project_id,

            "assigned_user_id": random.choice(
                company_users["user_id"].tolist()
            ),

            "task_name": random.choice(task_names),

            "task_status": status,

            "priority": random.choices(
                priority,
                weights=[20,50,25,5]
            )[0],

            "created_date": created_date.date(),

            "due_date": (
                created_date
                +
                timedelta(days=random.randint(7,60))
            ).date(),

            "completed_date": completed_date

        })


        task_id += 1
```

```python
tasks_df = pd.DataFrame(tasks)
tasks_df.head()
```

```python
tasks_df.to_csv(
    "projectflow-analytics/data/raw/tasks.csv",
    index=False
)
```
